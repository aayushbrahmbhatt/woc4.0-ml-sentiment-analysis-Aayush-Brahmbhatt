## Extracting Numbers From String

```python
ind = df['Unique ID'].apply(lambda x: x.split('_')[1])
df['ind'] = ind
df = df.astype({'ind': int })
```

 ## Indexing

```python
df = df.set_index('ind')
df
```

```python
df.drop('Unique ID',axis = 1,inplace = True)
```

```python
df.loc[1] # Index will start from 1
```

```python
df.iloc[0] # In python index starts from 0
```

## Dealing with NaN Values

```python
items = df.isna().sum() # Checking where in the data NaN value exist
items
```

```python
# Checking if the given column has NaN value and checking whether we can replace that NaN by mean of the other values in column
for item in items.iteritems():
    if df[item[0]].dtype != 'object' and item[1] > 0:
        df[item[0]].fillna(value=df[item[0]].mean(),inplace = True)
```

```python
df = df.dropna() # For deleting NaN object 
```

```python
df.isna().sum()
```

## Memory Optimisation

```python
# Converting datatype of YearBuilt from float64 to int32 because, Year can not be a float value so by changing the datatype we can reduce the space 
df = df.astype({'YearBuilt': int })
df['YearBuilt'].dtype
```

## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
 
# Encode labels in column 'Suburb' & 'Regionname'.
# Similarly we can do for other columns 
df['Suburb']= encoder.fit_transform(df['Suburb'])
df['Regionname']= encoder.fit_transform(df['Regionname'])
df.head()
```
